## 30 random points

```python
import random
import math
from typing import List, Tuple

def generate_random_points(num_points: int) -> List[List[float]]:
    """
    Generate random geographical points across the entire globe.
    
    Args:
        num_points: Number of random points to generate
    
    Returns:
        List of [lat, lon] coordinates representing random points on the globe
    """
    random_points = []
    
    for _ in range(num_points):
        # For truly uniform distribution on a sphere surface:
        # 1. Generate uniform random variables u, v in [0,1]
        # 2. lat = arcsin(2u - 1)
        # 3. lon = 2π*v
        
        u = random.random()  # Random value between 0 and 1
        v = random.random()  # Random value between 0 and 1
        
        # Convert to spherical coordinates
        lat = math.degrees(math.asin(2 * u - 1))  # Range: -90 to 90 degrees
        lon = math.degrees(2 * math.pi * v - math.pi)  # Range: -180 to 180 degrees
        
        random_points.append([lat, lon])
    
    return random_points
```

```python
a = generate_random_points(30)
```
